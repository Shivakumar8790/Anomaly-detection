```python
# Data Preprocessing
```

```python
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
import string

nltk.download("punkt")
nltk.download("stopwords")
```

```python
def text_preprocessing(text):
    text = text.lower()
    
    tokens = word_tokenize(text)
    # Remove punctuation and stopwords
    
    tokens = [word for word in tokens if word.isalnum()]
    
    tokens = [word for word in tokens if word not in stopwords.words("english")]
    
    return " ".join(tokens)
```

```python
final_doc = [text_preprocessing(doc) for doc in documents]
final_doc
```

```python
# Feature Extraction
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
x = vectorizer.fit_transform(final_doc)
```

```python
print(x)
```

```python
# Model training
```

```python
from sklearn.ensemble import IsolationForest
model = IsolationForest(n_estimators=100, contamination = 0.1)
```

```python
model.fit(x)
```

```python
# Anomaly detection
```

```python
IF_anomaly_scores = model.decision_function(x)
IF_anomaly_scores
```

```python
IF_anomalies = model.predict(x)

# -1 for anomalies, 1 for normal
IF_anomalies
```

```python
from sklearn.neighbors import LocalOutlierFactor

# Train LOF model
model_lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
```

```python
LOF_anomalies= model_lof.fit_predict(x)
```

```python
# Evaluation
```

```python
import pandas as pd
results = pd.DataFrame({"document":documents, "Final_doc":final_doc, 
                        "IF_anomaly_scores":IF_anomaly_scores, "IF_anomalies":IF_anomalies,
                        "LOF_anomalies":IF_anomalies})

results
```

```python
import matplotlib.pyplot as plt

# Count the number of anomalies for isolation forest
num_anomalies = len(results[results['IF_anomalies'] == -1])
num_normal = len(results[results['IF_anomalies'] == 1])

# Create a pie chart
labels = ['Anomalies', 'Normal']
sizes = [num_anomalies, num_normal]
colors = ['#ff9999', '#66b3ff']

plt.figure(figsize=(6, 4))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Anomaly Detection Results')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
